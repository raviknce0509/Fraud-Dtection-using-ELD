```python
%pip install folium 
%pip install openpyxl
import os
import numpy as np
import pandas as pd
import folium
```

```python

# read Excel file into a Pandas dataframe
df = pd.read_excel('DATA/TestData.xlsx')

# print the dataframe to verify it was imported correctly
df.head(5)

```

```python
#One approach
# create map object with center at the first location in the dataframe
m = folium.Map(location=[df['Latitude'][0], df['Longitude'][0]], zoom_start=10)

# add markers for each location in the dataframe
for index, row in df.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']]).add_to(m)

# display the map
m
```
